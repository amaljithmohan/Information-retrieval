```python
import nltk
import os
import nltk.corpus
nltk.download("punkt")
```

```python
data = (r"/content/book1.txt",r"/content/book2.txt",r"/content/book3.txt")
```

```python
data
```

```python
import string
```

```python
docs= []
for i in data:
  df = open(i, "r")
  r=df.read()
  remove_punct = r.translate(str.maketrans('', '', string.punctuation))
  docs.append(remove_punct.split())
print(docs)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopword = stopwords.words("english")
print(stopword)
```

```python
doc_tokens=[]
for i in range(0, len(docs)):
  doc_tokens.append(i*[])
  for word in docs[i]:
    token = word.lower()
    if token not in stopword:
      doc_tokens[i].append(token)
```

```python
print(doc_tokens)
```

```python
unique_words=[]
for i in range(0,len(docs)):
  for j in doc_tokens[i]:
    if j not in unique_words:
      unique_words.append(j)
```

```python
print(unique_words)
```

```python
def vector(token): 
  token_vec=[]
  for i in doc_tokens:
    if token in i:
      token_vec.append(1)
    else:
      token_vec.append(0)
  return token_vec
```

```python
vector('smart')
```

```python
vector_dict={}
for i in unique_words:
    vector_dict[i] = vector(i)

print(vector_dict)
```

```python
from nltk.tokenize import word_tokenize
```

```python
query=input("Enter the query:")
query=word_tokenize(query)
operator_words=[]
query_words=[]

for word in query:
    if word.lower() != "and" and word.lower() != "or" and word.lower() != "not":
      query_words.append(word.lower())
    else:
      operator_words.append(word.lower())


for word in operator_words:
  words_list1 = vector_dict[query_words[0]]
  words_list2 = vector_dict[query_words[1]]
  if word == "and":
    query_op = [w1 & w2 for (w1,w2) in zip(words_list1, words_list2)]
  elif word == "or":
    query_op = [w1 | w2 for (w1,w2) in zip(words_list1, words_list2)]
  elif word == "not":
    query_op = [not w1 for w1 in words_list2]
    query_op = [int(b == True) for b in query_op]
    query_op = [w1 & w2 for (w1, w2) in zip(words_list1, query_op)]
  
print(query_op)
```

```python
for i in list(zip(data,query_op)):
    if i[1]==1:
      print(i[0])
```

# B) Inverted Index

```python
def in_token(word):
  inverted=[]
  for i in range(0,len(doc_tokens)):
    if word in doc_tokens[i]:
      inverted.append(i)
  return inverted
```

```python
for i in unique_words:
  ind=in_token(i)
  print(i,':',ind)
```

```python
ind_dict={}
for i in unique_words:
    ind_dict[i] = {"docs":in_token(i)}

print(ind_dict)
```

```python
def doc_query(q):
  for i in q.split():
    if i=='and':
      q= q.replace(i,'&')
    elif i=='or':
      q=q.replace(i,'|')
    elif i=='not':
      q=q.replace(i,"^")
  for i in q.split():
    if i.isalpha():
      q=q.replace(i,repr(set(ind_dict[i]['docs'])))
  return q
```

```python
q=input("Enter the query :")
d=eval(doc_query(q))
if d==set():
  print("No documents")
else:
  for i in d:
    print("doc",i+1,end="")
```
