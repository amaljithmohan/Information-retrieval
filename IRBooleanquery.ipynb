```python
import string
```

```python
docs= []
for i in data:
  df = open(i, "r")
  r=df.read()
  remove_punct = r.translate(str.maketrans('', '', string.punctuation))
  docs.append(remove_punct.split())
print(docs)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopword = stopwords.words("english")
print(stopword)
```

```python
doc_tokens=[]
for word in docs:
  for i in range(0, len(word)):
    token = word[i]
    token = token.lower()
    if token not in stopword:
      doc_tokens.append(token)
```

```python
print(doc_tokens)
```

```python
unique_words=[]
for i in doc_tokens:
  if i not in unique_words:
    unique_words.append(i)
```
